```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Read in the CSV files
top_2019 = pd.read_csv("Resources/top10s.csv")
decades = pd.read_csv("Resources/1921_2020.csv")
```

```python
# Drop Extra Columns in the top19 df
dropped_df = top_2019.drop(columns = ['title', 'artist', 'top genre', 'Unnamed: 0'])
dropped_df
```

```python
# Rename Columns in top19 df
rename_top10 = dropped_df.rename(columns = {"bpm" : "tempo(BPM)",
                                         "nrgy" : "energy",
                                         "dnce" : "danceability",
                                         "dB" : "loudness",
                                         "live" : "liveness",
                                         "val" : "valence",
                                         "dur" : "duration",
                                         "acous" : "acousticness",
                                         "spch" : "speechiness",
                                         "pop" : "popularity"
                                        })
rename_top10
```

```python
# Convert Duration Column
ms_duration = rename_top10['duration']*1000
rename_top10['duration'] = ms_duration
```

```python
# Create a Function to Convert Columns to 0 - 1
def rescale(column):
    rescale_col = rename_top10[column]/100
    rename_top10[column] = rescale_col
    return rename_top10[column]
```

```python
# Rescale All Columns
columns_rescale = ['acousticness', 'danceability', 'energy', 'liveness', 'valence', 'speechiness']
for col in columns_rescale:
    rescale(col)
```

```python
# Drop Any Rows with NA values
rename_top10 = rename_top10.dropna(how = 'any')
rename_top10
```

```python
cleaned_top10 = rename_top10[['acousticness','danceability','duration','energy','liveness','loudness','speechiness','tempo(BPM)','valence','year','popularity']]
cleaned_top10
```

```python
cleaned_top10.to_csv("Resources/cleaned_top10.csv", index = False, header = True)
```
