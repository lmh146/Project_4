```python
# Dependencies
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

### Data Reading and Categorizing

```python
cleaned_top_10 = pd.read_csv("Resources/cleaned_top10.csv")
cleaned_top_10
```

```python
big_df = pd.read_csv("Resources/bigdf_binary.csv")
big_df
```

```python
big_df = big_df.loc[:, ['acousticness','danceability','duration','energy','liveness','loudness','speechiness','tempo(BPM)','valence','year','popularity']]
big_df
```

### Data Preprocessing

```python
# Scale Data
X = big_df[['acousticness','danceability','duration','energy','liveness','loudness','speechiness','tempo(BPM)','valence','year']]
y = big_df['popularity']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
data = [X_train_scaled, X_test_scaled, y_train, y_test]
```

```python
y_train
```

### SVC Model

```python
# Fit to the training data, and validate with the testing data
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
```

```python
# Plot the decision boundaries
x_min = X_train_scaled[:, 10].min()
x_max = X_train_scaled[:, 10].max()
y_min = X_train_scaled[:, 1].min()
y_max = X_train_scaled[:, 1].max()

# # XX, YY = np.mgrid[x_min:x_max, y_min:y_max]

# XX, YY = np.mgrid[X_train_scaled[0:10], X_train_scaled[0:1]]
# Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# # Put the result into a color plot
# Z = Z.reshape(XX.shape)
# # plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
# plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
#             linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
plt.scatter(X_train_scaled[:, 10], X_train_scaled[:, 1], c=y_train, cmap='bwr', edgecolor='k', s=100)
plt.show()
```

```python
# Calculate the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["unpopular", "popular"]))
```
